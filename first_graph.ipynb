```python
from typing import TypedDict, List
from langgraph.graph import StateGraph, START, END
```

```python
class AgenetState(TypedDict):
    name: str

def greeting(state: AgenetState) -> AgenetState:
    state['name'] = f"{state['name']}, you are doing amzing job learning LandGraph!"
    return state
```

```python
graph = StateGraph(AgenetState)
graph.add_node("greet_user", greeting)
graph.add_edge(START, "greet_user")
graph.add_edge("greet_user", END)

app = graph.compile()
```

```python
result = app.invoke(AgenetState(name="Animesh"))
  # Output: Animesh, you are doing amzing job learning LandGraph!
```
